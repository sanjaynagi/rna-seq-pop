```python
import matplotlib.pyplot as plt
from matplotlib_venn import *
import pandas as pd
import numpy as np
%matplotlib inline
```

### Venn Diagrams in python

```python
def plotvenn2(name, group1, group2, nboth,stat="PBS", group1name='Significant up DE genes', group2name='High PBS'):
    
    print(f"There are {group1} DE genes in {name}") 
    print(f"There are {group2} high Fst genes in {name}") 
    print(f"There are {nboth} shared in {name}")
    
    venn2(subsets = (group1, group2, nboth), set_labels = (group1, group2), 
          set_colors=('r', 'g'), 
          alpha = 0.5);
    venn2_circles(subsets = (group1, group2, nboth))
    plt.title(f"{name}")
    plt.savefig(f"../../results/venn/{name}_DE.{stat}.venn.png")
    plt.close()
```

```python
comparisons_path = "../../resources/DE.contrast.list"

comparisons = pd.read_csv(comparisons_path)
comparisons = comparisons.contrast.str.split("_", expand=True)
comparisons = [list(row) for i,row in comparisons.iterrows()]

pbscomps = [['FenAngola', 'ContAngola', 'NewOrleans'], ['FenAngola', 'ContAngola', 'Rockefeller']]

percentile = 0.05
```

```python
#### Differential expression v Fst venn diagram
for comp1,comp2 in comparisons:
    name = comp1 + "_" + comp2
    print(f"-------------- Venn Diagram for {name} --------------")
    de = pd.read_csv(f"../../results/genediff/{name}.csv")
    fst = pd.read_csv("../../results/variants/Fst_PBS.tsv", sep="\t")

    #compare sig DE genes and top 5% fst genes?
    #get sig up and down diffexp genes
    sigde = de[de['padj'] < 0.05]
    sigde_up = sigde[sigde['FC'] > 1]
    sigde_down = sigde[sigde['FC'] < 1]

    #take top 5% of fst genes
    highfst = fst.nlargest(int(fst.shape[0]*percentile),f"{name}_Fst")

    #how many fst? how many sig de up and down?
    nfst = highfst.shape[0]
    nde_up = sigde_up.shape[0]
    nde_down = sigde_down.shape[0]

    print(f"There are {nde_up} significantly upregulated genes in {name}") 
    print(f"There are {nde_down} significantly downregulated genes in {name}")

    intersection = [x for x in list(sigde_up.GeneID) if x in list(highfst.GeneID)]
    de_intersected = de[de.GeneID.isin(intersection)]
    nboth = de_intersected.shape[0]
    de_intersected.to_csv(f"../../results/venn/{name}_DE.Fst.intersection.tsv", sep="\t")
    
    plotvenn2(name, nde_up, nfst, nboth,stat="Fst", group1name='Significant up DE genes', group2name='High Fst')
```

### PBS

```python
#### Differential expression v Fst v PBS venn diagram
for pbscomp in pbscomps:
    name = pbscomp[0] + "_" + pbscomp[1] + "_" + pbscomp[2]
    name2 = pbscomp[2]+ "_" + pbscomp[1]
    
    de = pd.read_csv(f"../../results/genediff/{name2}.csv")
    pbs = pd.read_csv("../../results/variants/Fst_PBS.tsv", sep="\t")

    #compare sig DE genes and top 5% fst genes?
    #get sig up and down diffexp genes
    sigde = de[de['padj'] < 0.05]
    sigde_up = sigde[sigde['FC'] > 1]
    sigde_down = sigde[sigde['FC'] < 1]

    #take top 5% of fst genes
    highpbs = pbs.nlargest(int(pbs.shape[0]*percentile),f"{name}PBS")
    highfst = pbs.nlargest(int(pbs.shape[0]*percentile),f"{name2}_Fst")

    #how many fst? how many sig de up and down?
    nfst = highfst.shape[0]
    npbs = highpbs.shape[0]
    nde_up = sigde_up.shape[0]
    nde_down = sigde_down.shape[0]

    print(f"There are {nde_up} significantly upregulated genes in {name}") 
    print(f"There are {nde_down} significantly downregulated genes in {name}")
    
    #get intersection of PBS and Fst
    defstint = [x for x in list(sigde_up.GeneID) if x in list(highfst.GeneID)]
    nbothdefst = len(defstint)
    
    depbsint = [x for x in list(sigde_up.GeneID) if x in list(highpbs.GeneID)]
    nbothdepbs = len(depbsint)
    
    print(f"There are {npbs} high PBS genes in {name}") 
    print(f"There are {nboth} shared in {name}")

    venn2(subsets = (sigde_up.shape[0], npbs, nboth), set_labels = ('Significant DE genes', 'High PBS'), 
          set_colors=('r', 'g'), 
          alpha = 0.5);
    venn2_circles(subsets = (sigde_up.shape[0], npbs, nbothdepbs))
    plt.title(f"{name}")
    plt.savefig(f"../../results/venn/{name}_DE_PBS.venn.png")
    plt.close()
    
    ##intserct of fst and pbs
    defstint = [x for x in list(highpbs.GeneID) if x in list(highfst.GeneID)]
    nbothfstpbs = len(defstint)
    
    #get intersection of FST PBS and DE results 
    threesets = list(set(sigde_up.GeneID).intersection(highfst.GeneID, highpbs.GeneID))
    de_intersected3 = de[de.GeneID.isin(threesets)]
    nall = de_intersected3.shape[0]
    de_intersected3.to_csv(f"../../results/venn/{name2}_DE.Fst.PBS.intersection.tsv", sep="\t")
        
    venn3(subsets = (nde_up, nfst,nbothdefst , npbs, nbothdepbs , nbothfstpbs, nall), 
          set_labels = ('Significant DE', 'High Fst', 'High PBS'), 
          alpha = 0.5)
    venn3_circles(subsets = (nde_up, nfst,nbothdefst , npbs, nbothdepbs , nbothfstpbs, nall))
    plt.title(f"{name}")
    plt.savefig(f"../../results/venn/{name}_DE_Fst_PBS.venn.png")
    plt.close()
```

### Tajimas D scan

```python
tajimas_d = pd.read_csv("../../results/variants/tajimas_d.tsv", sep="\t")
```

```python
import allel
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tajimas_d.head()
```

```python
samples = pd.read_csv("../../config/samples.tsv", sep="\t")
chroms = [1,2,3]
```

```python
for chrom in chroms:
    for pop in samples.treatment.unique():
        tajdf = tajimas_d[tajimas_d.chrom == chrom]
        taj = allel.moving_mean(tajdf[f'{pop}_Tajima_d'], 10, 5)
        pos = allel.moving_mean(tajdf.Gene_midpoint, 10, 5)
        xtick = np.arange(0, pos.max(), 10000000)

        
        print(f"Mean Tajimas D across all genes on chromosome {chrom} for {pop} is {np.nanmean(tajdf[f'{pop}_Tajima_d'])}")
        plt.figure(figsize=[20,10])
        sns.lineplot(pos, taj)
        plt.xticks(xtick, rotation=45, horizontalalignment='right')
        plt.title(f"{pop} Tajimas D across chromosome {chrom}")
```

Sequence Diversity 

```python
seqdiv = pd.read_csv("../../results/variants/sequence_div.tsv", sep="\t")
```

```python
for chrom in chroms:
    for pop in samples.treatment.unique():
        seqdf = seqdiv[seqdiv.chrom == chrom]
        seq = allel.moving_mean(seqdf[f'{pop}_SeqDiv'], 50, 25)
        pos = allel.moving_mean(seqdf.Gene_midpoint, 50, 25)

        plt.figure(figsize=[20,10])
        sns.lineplot(pos, seq)
        plt.title(f"{pop} Sequence diversity across chromosome {chrom}")
```
