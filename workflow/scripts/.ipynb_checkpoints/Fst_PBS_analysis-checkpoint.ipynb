```python
def list_duplicates(seq):
    tally = defaultdict(list)
    for i,item in enumerate(seq):
        tally[item].append(i)
    return (tally)

def mean_pbs(ac1, ac2, ac3, window_size, normalise):
    
    pbs = allel.pbs(ac1, ac2, ac3, window_size=window_size, normed=normalise)
    meanpbs = np.nanmean(pbs)
    
    return()
```

### Filtering the input VCFs for quality

There are some crazy high quality values so to get a view from the data (by histogram), I just need to remove some outliers first before plotting. Ill use the function `is_outlier()`, located in `sweep_tools.py`.

Lets plot this is a histogram and see the distribution of quality scores.

```python
qual = vcf['variants/QUAL']
fltqual = qual[~is_outlier(qual)]

sns.distplot(fltqual)
```

Lets remove variants with a QUAL score below 30. We also need to take care of missingness: we expect high rates here due to this being RNA-Seq data, and therefore some treatments may not express genes entirely, eliminating the possibility of finding SNPs.

```python
passfilter = (vcf['variants/QUAL'] >= 30)
geno = geno.compress(passfilter, axis=0)
pos = pos.compress(passfilter, axis=0)
```

### Filter to res, sus comparison only

```python
comparison = dict((k, sampleIndices[k]) for k in ('FenAngola', 'ContAngola'))

res, sus = list(comparison.values())

res_gn = geno.take(res, axis=1)
sus_gn = geno.take(sus, axis=1)

gn_comparison = np.concatenate([res_gn, sus_gn], axis=1)
indices = [np.arange(15), np.arange(15, 30)]
```

### Missingness filter

```python
ac = allel.GenotypeArray(gn_comparison).count_alleles()
snpcounts = ac.sum(axis=1)
sns.distplot(snpcounts)
```

```python
fltprop = snpcounts.max()*(2/3) #must have at least 2/3rds alleles present
missingness_flt = snpcounts > fltprop
pos = allel.SortedIndex(pos[missingness_flt])

sus_ac = allel.GenotypeArray(sus_gn).count_alleles()
res_ac = allel.GenotypeArray(res_gn).count_alleles()

sus_ac = sus_ac.compress(missingness_flt, axis=0)
res_ac = res_ac.compress(missingness_flt, axis=0)
```
